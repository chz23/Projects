## Overview of Features Selection & EDA
---

The most rigorous aspect of this project, features selection was done in tandem with EDA for best results. Key aspects are as follows:

**External Research**

Used to justify the _inclusion_ of certain features, external research was crucial in determining not just what is generally important, but also how the _local_ context may affect relative importance.

It was also instrumental in explaining the strong presence of outliers, all of which belonged to the high-end side of things. A portion was dedicated to exploring these unusually expensive resale flats, which eventually justified the need for a separate model.

EDA was carried out by feature to verify whether the data supported the research.

**Dropping Collinear/Related Features**

With the large number of features, it was inevitable that several were closely related to one another. This segment was dedicated to identifying such features for removal. In addition their related features were named by way of explaining why the former were being dropped.

**Dropping Impractical/Non-Meaningful Features**

The second installment in the ongoing saga of aggressive trimming, features were singled out for either being too impractical to deal with or not pulling their weight. The former was dominated by categorical features that had high numbers of unique values such that they would be impractical to include in the modelling process, and that probably didn't provide much added value anyway - most contributed information that could be found in other features or a combination of features. 

Similar to the segment above, explanations were provided to justify removal.

**Dropping Features Based on EDA**

Features that survived the previous rounds but did not make the final cut were eliminated in this segment. EDA was carried out for these remaining features that had thus far flown under the radar, with the results being used to justify whether or not they got to stay.

**Summary**

To conclude, the features making up the respective models will be listed for convenience and recap purposes.

```python
# Function to create line charts.

def line(df=None, min_val=None, max_val=None, is_range=False, interval=None, 
         feature_1=None, feature_2=None, xlabel=None, ylabel=None, title=None, ylim=None):
    
    '''
    Parameters
    ----------
    df : pd.DataFrame
        DataFrame to be used in function.
    min_val : int
        Minimum y_axis value
    max_val : int
        Maximum y_axis value
    is_range: bool
        If True, x-axis data is by range.
    interval: bool
        Interval for x-axis range values.
    feature_1: str
        Feature to use for x-axis. 
    feature_2: str
        Feature to use for y-axis.
    xlabel: str
        x-axis label.
    ylabel: str
        y-axis label.
    title: str
        Title of the line graph.
    ylim: int
        Topmost label on y-axis.
    '''

    if is_range == True:
    
        # Create dictionary to store values for use in the line graph. 
        # Each key:value pair corresponds to an x-axis value and its corresponding y.
        # y-axis values are calculated by filtering the DataFrame using feature_1 and the corresponding range...
        # ...and obtaining the mean value.
        avg = {val: \
               np.round(np.mean(df[(df[feature_1] >= val) & (df[feature_1] < val+interval)][feature_2].values)) \
               for val in range(min_val, max_val, interval)}
        
    else:
            
        # Same as the above, except y-axis values are calculated by specific x values, not a range of values.
        avg = {val: \
               np.round(np.mean(df[df[feature_1] == val][feature_2].values)) \
               for val in df[feature_1].unique()}
        
    # Get rid of empty key:value pairs. 
    avg = {key: val for key, val in avg.items() if val > 0}

    # Create a line chart based on the above.
    chart = pd.DataFrame.from_dict(avg,
                                   orient='index', 
                                   columns=[f'avg_{feature_2}']).plot(y=f'avg_{feature_2}', 
                                                                      xlabel=xlabel,
                                                                      ylabel=ylabel,
                                                                      title=title,
                                                                      marker='o',
                                                                      legend=False)

    # Set topmost label on y-axis.
    chart.set(ylim = (0, ylim))
    ticks_loc = chart.get_yticks().tolist()
    chart.yaxis.set_major_locator(mticker.FixedLocator(ticks_loc))
    
    # If the topmost y-axis label exceeds 100,000, replace the last 3 0s with a 'K' denoting thousand.
    if ylim > 100_000:
        chart.set_yticklabels(['{:,.0f}'.format(y/1000) + 'K' for y in ticks_loc]);
        
    else:
        chart.set_yticklabels(['{:,.0f}'.format(y) for y in ticks_loc])
```

```python
# Function to create 2 line overlaying line graphs.

def compare_lines(df=None, min_val=None, max_val=None, is_range=False, interval=None, 
                  feature=None, xlabel=None, title=None, ylim=None, ylim_2=None):

    '''
    Parameters
    ----------
    df : pd.DataFrame
        DataFrame to be used in function.
    min_val : int
        Minimum y_axis value
    max_val : int
        Maximum y_axis value
    is_range: bool
        If True, x-axis data is by range.
    interval: bool
        Interval for x-axis range values.
    feature: str
        Feature to use for x-axis. 
    xlabel: str
        x-axis label.
    ylabel: str
        y-axis label.
    title: str
        Title of the line graph.
    ylim: int
        Topmost label on y-axis (left).
    ylim_2: int
        Topmost label on y-axis (right).
    '''

    if is_range == True:
    
        # Create dictionary to store values for use in the line graph. 
        # Each key:value pair corresponds to an x-axis value and its corresponding...
        # ...2-elements list of y values for each of the line graphs.
        # y-axis values are calculated by filtering the DataFrame using feature_1 and the corresponding range...
        # ...and obtaining two things: the mean resale price, and the mean price per sqm.
        avg = {val: [
            np.round(np.mean(df[(df[feature] >= val) & (df[feature] < val+interval)].resale_price.values)), 
            np.round(np.mean(df[(df[feature] >= val) & (df[feature] < val+interval)].resale_price.values) \
                     / np.mean(df[(df[feature] >= val) & (df[feature] < val+interval)].floor_area_sqm.values))
        ] for val in range(min_val, max_val, interval)}
        
    else:
        
        # Same as the above, except y-axis values are calculated by specific x values, not a range of values.
        avg = {val: [
            np.round(np.mean(df[df[feature] == val].resale_price.values)), 
            np.round(np.mean(df[df[feature] == val].resale_price.values) \
                     / np.mean(df[df[feature] == val].floor_area_sqm.values))
        ] for val in df[feature].unique()}
        
    # Get rid of empty key:value pairs. 
    avg = {key: [val[0], val[1]] for key, val in avg.items() if (val[0] > 0) and (val[1] > 0)}
    
    # Save the above into a DataFrame.
    df_chart = pd.DataFrame.from_dict(avg, orient='index', columns=[f'avg_resale_price', 'avg_price_per_sqm'])

    # Create a line graph based on the above.
    chart = df_chart.plot(y=f'avg_resale_price', 
                          xlabel=xlabel, 
                          ylabel='Mean Resale Price ($)', 
                          title=title, 
                          marker='o', 
                          legend=False)

    # Create a duplicate of the line graph.
    chart_2 = chart.twinx()

    # Add the second line graph in and plot the values accordingly.
    df_chart.plot(y='avg_price_per_sqm', 
                  ax=chart_2, legend=False, 
                  color='r', 
                  ylabel='Mean Price per sqm ($)', 
                  style='o-')

    # Add a legend, since there are two different lines.
    chart.figure.legend(['Mean Resale Price', 'Mean Price per sqm'], loc='upper right')
    
    # Set topmost label on y-axis (left).
    chart.set(ylim = (0, ylim))
    ticks_loc = chart.get_yticks().tolist()
    chart.yaxis.set_major_locator(mticker.FixedLocator(ticks_loc))

    # Replace the last 3 0s with a 'K' denoting thousand.
    chart.set_yticklabels(['{:,.0f}'.format(y/1000) + 'K' for y in ticks_loc])
        
    # Set topmost label on y-axis (right).
    chart_2.set(ylim = (0, ylim_2))
    ticks_loc_2 = chart_2.get_yticks().tolist()
    chart_2.yaxis.set_major_locator(mticker.FixedLocator(ticks_loc_2))

    chart_2.set_yticklabels(['{:,.0f}'.format(y) for y in ticks_loc_2])
    
    plt.figure(figsize=(12,8))
    plt.show();
```

```python
def bar(df=None, feature_1=None, feature_2=None, ppsqm=False, x_axis_labels=None,
        xlabel=None, ylabel=None, title=None, colour='tab:blue', figsize=(16,9), ylim=None):
    
    
    '''
    Parameters
    ----------
    df : pd.DataFrame
        DataFrame to be used in function.
    feature_1: str
        Feature to use for x-axis. 
    feature_2: str
        Feature to use for y-axis.
    ppsqm: bool
        If True, y-axis will be price per sqm.
    x_axis_labels: list of str
        Labels to use for x-axis.
    xlabel: str
        x-axis label.
    ylabel: str
        y-axis label.
    title: str
        Title of the line graph.
    colour: str, list
        Colour(s) to use for bars.
    figsize: tuple
        Size of figure.
    ylim: int
        Topmost label on y-axis.
    '''
    
    if ppsqm == True:
        
        # Create dictionary to store values for use in the line graph. 
        # Each key:value pair corresponds to an x-axis label and its corresponding y.
        # y-axis values are calculated by filtering the DataFrame using feature_1 and the corresponding range...
        # ...and obtaining the mean price per sqm.
        avg = {val: np.round(
            np.mean(df[df[feature_1] == val].resale_price.values) / np.mean(df[df[feature_1] == val].floor_area_sqm.values)
        ) for val in x_axis_labels}
        
    else:
        
        # Same as the above, except y-axis values are based on feature_2 rather than mean price per sqm.
        avg = {val: np.round(
            np.mean(df[df[feature_1] == val][feature_2].values)
        ) for val in x_axis_labels}
    
    # Create a bar graph based on the above.
    chart = pd.DataFrame.from_dict(avg,
                                   orient='index', 
                                   columns=[f'avg_{feature_2}']).plot(kind='bar',
                                                                      y=f'avg_{feature_2}', 
                                                                      xlabel=xlabel,
                                                                      ylabel=ylabel,
                                                                      title=title,
                                                                      legend=False,
                                                                      color=colour,
                                                                      figsize=figsize)
    
    # Set topmost label on y-axis.
    chart.set(ylim = (0, ylim))
    ticks_loc = chart.get_yticks().tolist()
    chart.yaxis.set_major_locator(mticker.FixedLocator(ticks_loc))
    
    # If the topmost y-axis label exceeds 100,000, replace the last 3 0s with a 'K' denoting thousand.
    if ylim > 10_000:
        chart.set_yticklabels(['{:,.0f}'.format(y/1000) + 'K' for y in ticks_loc]);
    else:
        chart.set_yticklabels(['{:,.0f}'.format(y) for y in ticks_loc])
```

```python
# Function that takes a given feature and returns various information...
#  ...(both categorical & numerical features) specific to it.

def features_analysis(df=None, feature=None, n_extras={}):
    
    '''
    Parameters
    ----------
    df : pd.DataFrame
        DataFrame to be used in function.
    feature: str
        Feature to return information on.
    n_extras: dict, {str: str}
        Dictionary where each key is a feature and the corresponding value must be
        either 'categorical' or 'numerical'.
    '''
    
    # Create a dictionary to store the average resale price for each label of the feature.
    d = {
        label: 
        [df[df[feature] == label].shape[0], 
         np.round(df[df[feature] == label].resale_price.mean())] 
        for label in df[feature].unique()
    }

    for extra_feature in n_extras.keys():

        if n_extras[extra_feature] == 'categorical':
            
            # For categorical features, for each label of the main feature...
            # ...include all corresponding labels of the categorical feature.
            for label in d.keys():
                d[label].append(df[df[feature] == label][extra_feature].unique())

        elif n_extras[extra_feature] == 'numerical':
            
            # For numerical features, for each label of the main feature...
            # ...calculate the mean value of the numerical feature.
            for label in d.keys():
                d[label].append(np.mean(df[df[feature] == label][extra_feature].mean()))
    
    # Create a list to store column names for the resulting DataFrame.
    cols = ['n_flats', 'mean_price'] + [header for header in n_extras.keys()]
                
    # Create a DataFrame to store the above.
    df_features = pd.DataFrame.from_dict(
        d,
        orient='index', 
        columns=cols
    )
    
    # If 'floor_area_sqm' is specified, include an additional column reflecting the average price per sqm.
    if 'floor_area_sqm' in n_extras.keys():
        df_features['price_per_sqm'] = df_features.loc[:, 'mean_price'] / df_features.loc[:, 'floor_area_sqm']
        
    return df_features.sort_values('price_per_sqm', ascending=False)
```

```python
# Function that accepts 2 levels of filtering and returns information (only for numerical features) specific to it.

def spec_price(df=None, feature_1=None, f1_label=None, feature_2=None, return_features=[]):
    
    '''
    Parameters
    ----------
    df : pd.DataFrame
        DataFrame to be used in function.
    feature_1: str
        First feature to filter information on.
    f1_label: str
        Specific label, from feature_1, to return information on.
    feature_2: str
        Second feature to filter information on.
    return_features: list
        Features to return. Must be numerical.
    '''
    
    # Filtered DataFrame based on feature_1 and its corresponding f1_label.
    df_spec = df[df[feature_1] == f1_label]
    
    # Create a dictionary to store, for each label in df_spec filtered by feature_2, its number of occurences.
    d = {label: [df[(df[feature_1] == f1_label) & (df[feature_2] == label)].shape[0]] \
         for label in df_spec[feature_2].unique()}

    for feature in return_features:
        
        # For each label in the dictionary, append its mean value.
        for label in d.keys():
            d[label].append(
                np.round(df[(df[feature_1] == f1_label) & (df[feature_2] == label)][feature].mean())
            )
    
    # Create a list to store column names for the resulting DataFrame.
    cols = [f'n_flats({f1_label})'] + [f'mean_{feature}' for feature in return_features]
    
    # Create a DataFrame from the above.
    df_features = pd.DataFrame.from_dict(
        d,
        orient='index', 
        columns=cols
    )
        
    # If 'floor_area_sqm' is specified, include an additional column reflecting the average price per sqm.
    if 'floor_area_sqm' in return_features:
        df_features['price_per_sqm'] = df_features.loc[:, 'mean_resale_price'] / \
        df_features.loc[:, 'mean_floor_area_sqm']
        
    return df_features.sort_values('price_per_sqm', ascending=False)
```

```python
# Function to determine the most characteristic features of flats belonging to...
#  ...the top and bottom 2.5% of flats by the specified filter.

def calc_proportion(df=None, col=None, sort_by=None, outlier=None, quantile=None):
    
    '''
    Parameters
    ----------
    df : pd.DataFrame
        DataFrame to be used in function.
    col: str
        Main feature that resulting DataFrame is based on.
    sort_by: str
        Additional feature to sort DataFrame by.
    outlier: str
        Either 'high' or 'low'.
    quantile: float
        Float that is larger than 0 and smaller than 1.
    '''
    
    # Create a DataFrame using the sort_by and quantile features to filter df.
    if outlier == 'low':
        
        df_outliers = df[df[sort_by] <= df[sort_by].quantile(q=quantile)]
        
    elif outlier == 'high':
        
        df_outliers = df[df[sort_by] >= df[sort_by].quantile(q=quantile)]

    df_grouped = df.groupby(col)[sort_by].describe().sort_values('mean')
        
    # Create a dictionary that takes the feature and extracts its labels to be the key...
    # ...while the values are a 2-element list comprising the label's number of occurences...
    # ...as well as its mean.
    info_by_feature = {feature: [count, mean] for feature, count, mean in \
                       zip(df_grouped.index, df_grouped['count'], df_grouped['mean'])}
        
    df_outliers = df_outliers.groupby(col)[sort_by].describe().sort_values('mean')
    
    # Same as the info_by_feature dictionary, but for the outliers.
    outliers_info = {feature: [count, mean] for feature, count, mean in \
                     zip(df_outliers.index, df_outliers['count'], df_outliers['mean'])}
    
    # Create a dictionary that takes each label of the feature to be the key and...
    # ...the value refers to the proportion of its occurences that are in df_outliers.
    proportion = {feature: 
                  np.round(outliers_info[feature][0]/info_by_feature[feature][0]*100, 2) \
                  for feature in outliers_info.keys()}
    
    # Return a DataFrame that shows, for each label of the feature, its proportion...
    # ...that is found in df_outliers.
    return pd.DataFrame(
        proportion.values(), 
        index=proportion.keys(), 
        columns=['proportion_in_percentage']
    ).sort_values('proportion_in_percentage', ascending=False)
```

## External Research
---

### Floor Area

_Relevant features: `flat_type`, `floor_area_sqm`, `floor_area_sqft`_

Floor area is usually a very important determinant of housing price, and this is especially true of tiny Singapore which, for better or worse, is amongst the densest countries in the world. It is expected that floor area is positively correlated with resale price.

### Remaining Lease Period

_Relevant features: `lease_commence_date`, `hdb_age`_

All HDB flats are subject to a 99-year lease period without exception. This likely impacts valuation prices and may even be a point of negotiation between prospective buyers and sellers. It is expected that remaining lease period is inversely correlated with resale price.

### Location

_Relevant features: `town`, `street_name`, `address`, `postal`, `Latitude`, `Longitude`, `planning_area`, `mrt_name`, `mrt_latitude`, `mrt_longitude`, `bus_stop_latitude`, `bus_stop_longitude`_

Although Singapore is modest in size, proximity to central areas still matters. Flats located at or closer to the city centre tend to be more expensive. The most obvious reason would be land valuation, in that such areas tend to be seen as prime areas with corresponding high land valuations. 

From the buyer's point of view, practical reasons for higher prices include connectivity (multiple train lines, good selection of buses, etc.), proximity to the CBD (where many people work at) and selection of amenities (many shopping malls, meal options, etc.). 

Finally, in general, regardless of country, such houses also tend to be viewed as more prestigious, and may automatically fetch higher prices. Thus, it is expected the more centrally located a flat is, the higher the resale price.

### Connectivity & Amenities

_Relevant features: `commercial`, `market_hawker`, `multistorey_carpark`, `precinct_pavilion`, `Mall_Nearest_Distance`, `Mall_Within_500m`, `Mall_Within_1km`, `Mall_Within_2km`, `Hawker_Nearest_Distance`, `Hawker_Within_500m`, `Hawker_Within_1km`, `Hawker_Within_2km`, `hawker_food_stalls`, `hawker_market_stalls`, `mrt_nearest_distance`, `bus_interchange`, `mrt_interchange`,`bus_stop_nearest_distance`, `pri_sch_nearest_distance`, `sec_sch_nearest_dist`_

For the average buyer, amenities probably do factor into decision-making, but the extent depends on type of amenity and the buyer themself. For example, while the presence of pavillions is unlikely to matter at all, proximity to the nearest MRT station probably does - especially for those who don't plan on driving everywhere. In particular, specific to the local context, proximity to either bus stops and/or MRT stations should be at least a point of consideration since commuting via public transport is very common.

It should be noted that several features in this dataset are concerned with whether or not the amenities exist _in the target block_, which is not particularly helpful. For example, for someone who places an emphasis on having convenient access to a minimart or supermarket, it matters how _close_ the the amenity is, not whether or not it is situated right in the target block.

### Putting It All Together

Individually, with all other things remaining constant, the abovementioned features may have strong correlations with resale prices. However, since resale price takes into account all of them (and more!), individual correlation values may not be _that_ strong due to the interaction of features. How does a spacious 5-room flat an hour away from the city measure up to a tiny 3-room flat within walking distance of the CBD district? How do prospective buyers _rank_ these features when looking at resale flats? Conducting EDA might be a good way of shedding some light on these questions.

### Answer:

Correlation is weaker than `floor_area_sqm` vs `resale_price`. At most, this tells us that prospective buyers are more mindful of floor area than remaining lease period. In other words, it doesn't mean that it's not important - it's just less important _when compared_ to floor area. 

The blue line, `hdb_age_at_tranc` vs. **mean** resale price, shows an obvious trend suggesting that the older the resale flat, the lower its resale price. This is contrasted with the red line, where price per sqm is used instead of resale price, and it seems that there's actually a range of values where average price per sqm actually _increases_ with age. 

Locally, there's a plausible explanation for this that may also serve as an insight into why the correlation between age and price is not that strong - flat sizes are getting smaller and smaller with each passing year, with the very largest flats of the past having been phased out entirely. The numbers match - most of these flats were phased out anywhere between 10 and 30 years ago, meaning that their age as at transaction fits nicely with the upwards trend seen in the chart. In other words, there exists a buyer profile where the desire to obtain a larger flat is outweighs the hesitance of getting an older flat. 

Note that `lease_commence_date`, `year_completed` & `hdb_age` can now be dropped in favour of the newly-created `hdb_age_at_tranc`.

```python
# Obtain correlation values.

pd.DataFrame(trn[['resale_price', 
                  'Mall_Nearest_Distance', 
                  'Hawker_Nearest_Distance', 
                  'mrt_nearest_distance',
                  'pri_sch_nearest_distance', 
                  'sec_sch_nearest_dist']].corr().iloc[1:, :1]).sort_values('resale_price', ascending=False)
```

```python
# Using all possible combinations of the 5 features, create the combination that...
# ...results in the best correlation with resale price.

corr_dict = {f'{num}_features': [0, []] for num in range(2, 6)}

for num in range(2, 6):
    
    combis = [list(i) for i in list(combinations(['sec_sch_nearest_dist', 
                                                  'pri_sch_nearest_distance', 
                                                  'Mall_Nearest_Distance', 
                                                  'Hawker_Nearest_Distance', 
                                                  'mrt_nearest_distance'], 
                                                 num))]

    for features in combis:

        trn['nearest_dist_amenities'] = [sum(i) for i in trn.loc[:, features].values]

        df_amenities_corr = trn[['resale_price', 'nearest_dist_amenities']].corr()
        
        # If correlation exceeds the current entry, it will replace the latter.
        if df_amenities_corr.iloc[1, -1] > corr_dict[f'{num}_features'][0]:
            corr_dict[f'{num}_features'][0] = df_amenities_corr.iloc[1, 0]
            corr_dict[f'{num}_features'][1] = features
            
trn.drop('nearest_dist_amenities', axis=1, inplace=True)
```

```python
# Obtain optimal correlation values.

pd.DataFrame.from_dict(corr_dict, orient='index', 
                       columns=['corr', 'features_combination']).sort_values('corr', ascending=True)
```

#### Pinnacle@Duxton

Fairly new and located centrally (Outram), Pinnacle@Duxton was built specifically to _stand out_ and has successfully done so, winning awards for its innovative design and impressive facilities, amongst other things. These highly coveted flats are stylish, statement properties and bear testament to what public housing in Singapore can be like at its best.

#### HDB lofts: the upgraded version of the discontinued Maisonette

HDB `Premium Apartment Lofts` are the answer to the discontinued maisonette. Despite that, they are still rather scarce because these lofts aren't frequently built, and often only in small amounts. They are also 'designer' estates that stand out from other HDB estates and have more arguably more practical layouts compared to maisonettes. A new, two-storey flat can easily be designed to look polished and upscale, which might explain why buyers are willing to shell out such high amounts (still lower than actual private properties). 

#### Once Upon a Landed House

HDB `Terrace` houses resemble actual terrace houses in that they are _actually landed properties_, but built by HDB. Here's the catch - they still come with the usual 99-year lease, meaning that flat owners **_do not own_** the land their house stands upon. They are slightly larger than typical 5-room flats (despite being, technically, 3-room and 4-room flats), rare (only 285 units exist, with 95 being recorded in this dataset) and the oldest of all flat models with an average age of 45 years as at the date of transaction. They are not being produced anymore.

For units with less than 50 years left on the lease, buyers are surprisingly willing to shell out a pretty penny for these flats. Why? It's likely a combination of location (central areas - Novena & Queenstown, with some units being walking distance to the MRT station) and the feeling of living in a landed property, albeit a rather small one. 

#### `DBSS` flats

Phased out in 2012, these Design, Build and Sell Scheme (DBSS) flats were marketed as higher-end flats as they were built by private developers, thus enabling them to shed the cookie-cutter image/design that your run-of-the-mill HDB estate has. They are made up mostly of 4-room and 5-room flats.

In particular, 5-room DBSS flats, which account for half of all DBSS flats, _actually_ deliver on the number of rooms - 4, as opposed to 3 (+ a study space) in typical 5-room flats. Case in point, across the 10 types of 5-room flats available in this dataset, DBSS 5-room flats come in at 3rd highest in terms of resale price ($838K), losing only to Pinnacle@Duxton flats and Premium Apartment Lofts. However, this does not mean that they are bigger - they are actually the 2nd smallest of all 5-room flats - but it does mean that home owners don't have to pump in additional money to convert the study into an actual room via addition of walls.

### Does it matter whether or not the flat's nearest MRT station is also a bus interchange?

```python
# What percentage of the dataset is comprised of flats whose nearest MRT station...
# ...also includes a bus interchange?

np.round(trn[trn.bus_interchange == 1].shape[0] / trn.shape[0]*100, 2)
```

```python
# This code shows the following information by region:
# 1) percentage of flats where nearest mrt station is also a bus interchange
# 2) average resale price of flats whose nearest mrt station is also a bus interchange
# 3) average resale price of flats who nearest mrt station is not a bus interchange

for region, region_str in zip([north, west, east, northeast, central], ['north', 'west', 'east', 'northeast', 'central']):
    
    n_0 = 1
    n_1 = 0
    price_0 = 0
    price_1 = 0
    
    for p_area in region:
        
        n_0 += trn[(trn.bus_interchange == 0) & (trn.planning_area == p_area)].shape[0]
        n_1 += trn[(trn.bus_interchange == 1) & (trn.planning_area == p_area)].shape[0]
        
        price_0 += sum(trn[(trn.bus_interchange == 0) & (trn.planning_area == p_area)].resale_price)
        price_1 += sum(trn[(trn.bus_interchange == 1) & (trn.planning_area == p_area)].resale_price)
        
    print(f'{region_str}: {np.round(n_1/trn.shape[0]*100)}%, ${np.round(price_1/n_1/1000)}K vs ${np.round(price_0/n_0/1000)}K')
```

### Answer:

At first glance, with more than half of all flats having their nearest MRT station also function as a bus interchange, this features seems redundant. However, a closer look at region reveals that it actually _does_ matter - at least for 3 out of 5 regions. 

### Does it matter whether or not the flat's nearest MRT station consists of multiple lines?

```python
# What percentage of the dataset is comprised of flats who nearest MRT station consists of multiple train lines?

np.round(trn[trn.mrt_interchange == 1].shape[0]/trn.shape[0]*100, 2)
```

```python
# Create a dictionary that shows, for each mrt interchange, which planning areas is it considered to be a part of.

resale_price_by_mrt_interchange = {
    mrt: {p_area: [] 
          for p_area in list(trn[trn.mrt_name == mrt].planning_area.unique())
         } for mrt in trn[trn.mrt_interchange == 1].mrt_name.unique()}
```

```python
# For each mrt interchange and its corresponding planning area(s), calculate the following:
# mean resale price of flats where it is the nearest station,
# mean resale price of other flats,
# difference between the former and the latter.

for mrt, p_area in resale_price_by_mrt_interchange.items():
    
    for key in p_area.keys():
        
        p_area[key].append(np.round(trn[(trn.planning_area == key) & \
                                        (trn.mrt_name == mrt)].resale_price.mean()))
        
        p_area[key].append(np.round(trn[(trn.planning_area == key) & \
                                        (trn.mrt_name != mrt)].resale_price.mean()))
        
        p_area[key].append(p_area[key][0] - p_area[key][1])
```

```python
# How does resale price vary according to the above information?

count = 0

for mrt, info in resale_price_by_mrt_interchange.items():
    for p_area, lst in info.items():
        if lst[2] > 0:
            count += 1
            
print(f'Out of {len(resale_price_by_mrt_interchange.keys())} planning areas that \
contain MRT interchanges, {count} have a higher mean resale price for flats whose \
nearest MRT station is the interchange.')
```

### Answer:

Given the high standards of Singapore's rail connectivity, this feature is rather debatable - would staying at, for example, Paya Lebar, make a buyer willing to pay a premium as opposed to staying a stop away at Eunos? They are only two minutes apart by train.

While some people might be willing to shell out extra to shave off the extra time needed to change trains, it is rather likely that there are _other_ factors involved as well - MRT interchanges tend to have many amenities in the area ranging from malls to bustling heartland scenes. Some of these factors might overlap with other features, but the exact mix is probably quite unique, hence it would probably be prudent to retain the `mrt_interchange` feature.
